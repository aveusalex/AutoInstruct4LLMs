```python
import pandas as pd
import os
```

```python
cwd = os.getcwd().split("AutoInstruct4LLMs")[0] + "AutoInstruct4LLMs"
path = f"{cwd}/dados/gold/demandas_identificadas/demandas_identificadas_v2.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.info()
```

```python
# verificando onde os valores estão como INVALID
df_invalid = df[df["fala_reescrita"] == "INVALID"]
```

```python
# verificando quantos valores estão como INVALID e a porcentagem
print(f"Quantidade de valores INVALID: {len(df_invalid)}")
print(f"Porcentagem de valores INVALID: {len(df_invalid)/len(df)*100:.2f}%")
```

```python
# verificando onde os valores não estão como INVALID
df_valid = df[df["fala_reescrita"] != "INVALID"]
```

```python
# verificando quantos valores não estão como INVALID e a porcentagem
print(f"Quantidade de valores não INVALID: {len(df_valid)}")
print(f"Porcentagem de valores não INVALID: {len(df_valid)/len(df)*100:.2f}%")
```

```python
# VERIFICANDO ALGUNS EXEMPLOS DE FALAS REESCRITAS
df_valid.sample(15)
```

```python
# iterando pelas linhas do dataframe printando a fala do cliente e a fala reescrita
for index, row in df.iterrows():
    print(f"Cliente: {row['fala_cliente']}")
    print(f"Reescrita: {row['fala_reescrita']}")
    print("--------------------------------------------------\n")
```

```python
df_teste1 = pd.read_csv(f"{cwd}/dados/gold/demandas_identificadas/demandas_identificadas_teste1.csv")
df_teste2 = pd.read_csv(f"{cwd}/dados/gold/demandas_identificadas/demandas_identificadas_teste2_promptunico.csv")
df_teste3 = pd.read_csv(f"{cwd}/dados/gold/demandas_identificadas/demandas_identificadas_teste3_newprompt.csv")

# verificando a quantidade de invalid em cada um
print(f"Quantidade de valores INVALID no teste1: {len(df_teste1[df_teste1['fala_reescrita'] == 'INVALID'])}")
print(f"Quantidade de valores INVALID no teste2: {len(df_teste2[df_teste2['fala_reescrita'] == 'INVALID'])}")
print(f"Quantidade de valores INVALID no teste3: {len(df_teste3[df_teste3['fala_reescrita'] == 'INVALID'])}")

# fazendo join dos dois dataframes pela coluna arquivo
df_test = df_teste1.join(df_teste2.set_index("arquivo"), on="arquivo", rsuffix="_teste2").join(df_teste3.set_index("arquivo"), on="arquivo", rsuffix="_teste3")
```

```python
aux = df_test[df_test.fala_reescrita == "INVALID"]
```

```python
for index, row in aux.iterrows():
    print(f"Cliente: {row['fala_cliente']}")
    print(f"Reescrita: {row['fala_reescrita']}")
    print(f"Reescrita teste3: {row['fala_reescrita_teste3']}")
    print("--------------------------------------------------\n")
```
